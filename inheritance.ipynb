```python
#single inheritance
# Parent Class
class Employee:
    def __init__(self):
        self.name = "divya"
        self.salary=30000

# Child Class
class child(Employee):
    def show(self):
        print("Name:", self.name)
        print("Salary:", self.salary)

employee = child()
employee.show()
```

```python
#multiple inheritance
#parent class
# Parent class 1
class Camera:
    def take_photo(self):
        print("Camera is taking a photo")


# Parent class 2
class Phone:
    def make_call(self):
        print("Phone is making a call")


# Child class
class Smartphone(Camera, Phone):
    def browse_internet(self):
        print("Smartphone is browsing internet")


s = Smartphone()
s.take_photo()
s.make_call()
s.browse_internet()
```
